Visualize is there is any missing values in the dataset based on a heat map

```python
#missing values
sns.heatmap(df.isnull(),cbar=False,cmap="viridis")
```

Find out the correlation betweeb variables in the given dataset correlation between different variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap="Blues")
```

Analyse how the income of a patient affects the no of visits to the hospital

```python
#relation between income and visits
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

Count and visualize the number of males and females affected by illness

```python
#No: of male and female affected by illness
sns.histplot(df.gender,bins=2)
```

Visualize the percentage of people getting govt health insurance due to low income due to old age and also percentage of people having private health insurance

```python
# % of people getting govt Insurance due to low income
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of people getting govt Insurance due to low income ")
plt.show()
# % of people having private Insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of people having private Insurance ")
plt.show()
# % of people getting govt health Insurance due to old age, disability or veterab status
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health Insurance due to old age, disability or veterab status  ")
plt.show()
```

Plot a horizontal bar chart to analyze the reduced days of activity due to illness based on Gender

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#creating the bar chart
plt.barh(db['gender'],db['reduced'],color = ['cornflowerblue','lightseagreen'])
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
#Show the plot
plt.show()
```
